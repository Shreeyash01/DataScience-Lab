```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
df.drop('deck',inplace=True,axis = 1)
df
```

```python
df.dropna(subset=['age'],inplace=True)
df
```

```python
df.describe()
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df,hue='survived')
```
