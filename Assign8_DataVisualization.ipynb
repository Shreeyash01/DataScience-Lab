```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
#dropping the column because it has too many null values
df.drop('deck',inplace=True,axis = 1)
df
```

```python
df.isnull().sum()
```

```python
#saving the data frame in other frame and doing things or preprocessing on the newly created dataset
#x1 is the new data frame and preprocessing is being done on x1
#df as it is safe rahega
x = df
x1 = df.dropna()
```

```python
x1.isnull().sum()
```

```python
sns.scatterplot(x='age',y='fare',data=x1) #this helps is plotting the plot that gives the information in 2d
```

```python
sns.scatterplot(x='age',hue = 'sex', y='fare',data=x1) #hue helps in giving colours and simplifying it to understand better
```

```python
sns.displot(x1['fare'],aspect = 2 )
```

```python
sns.displot(x1['fare'],aspect = 4)
```

```python
sns.displot(x1['fare'],aspect = 6) #increase in aspect attribute helps in broading the columns as shown in graph
```

```python
sns.displot(x1['fare'],kde = True)
```

```python
sns.displot(x1['fare'],kde = True,bins = 10) #bins help to divide the graph into 10 parts
```

```python
sns.displot(x1['age']) #helps in finding out the concentrated age of a praticular age group
```

```python
sns.displot(x1['age'],hue = 'sex') #hue will not run because it mostly works on 2d data and here we are performing action on single data
```
