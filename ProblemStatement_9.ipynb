```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.drop(['class','who','adult_male','alive','alone','deck','embark_town'],axis=1,inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
sns.boxplot(x='survived',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df,hue='survived') 
```

## Observations/Inference
- The box plot provides information about the distribution of age for each gender, split by survival status.
- The boxes represent the interquartile range (IQR), where the line in the middle of the box represents the median age.
- The whiskers extend to the minimum and maximum values within 1.5 times the IQR.
- Outliers, represented by individual points, are data points that fall outside the whiskers.

```python
sns.boxplot(x='sex',y='age',data=df,hue='pclass') 
```
